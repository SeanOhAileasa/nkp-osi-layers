###### 1. The Open Systems Interconnect (OSI) Model
***

Now in this presentation, we will introduce the OSI model, and that stands for Open Systems Interconnect. This is an architectural approach to developing network communications. [These layers are: Layer 1, Physical; Layer 2, Data Link;  Layer 3, Network;  Layer 4, Transport; Layer 5, Session; Layer 6, Presentation; Layer 7, Application.] You can see that there are seven layers. And these divide up the tasks so that each layer is responsible for only a sub-component of the overall communication process. This helps to facilitate consistency when developing new applications, new protocols, or new devices. 

So why does it even exist at all? Well, imagine you're a company, and in fact the only company in the entire world that developed anything and everything to do with computing. If that was the case, you'd certainly divide up your company into departments. And you'd assign each department a special area of focus. So that your application developers just don't have to concern themselves with the department that makes physical cabling and network adapters, hardware for example. Because if you did, if the application developers had to take that into account. Then they would have to develop one version of the application for one type of hardware. And then another version of the application for a different type of hardware or a different type of network media. So clearly, we don't want to have to do any of that. As an application developer, all I want to do is develop my application. I don't want to concern myself with anything in the lower layers.The network protocols, the physical hardware, I simply want to focus on the application. So that is really the overall approach. Now clearly, we have any number of companies that are developing these things, any number of individuals. All kinds of different types of applications, different protocols, different types of hardware, that all have to be able to ultimately communicate with each other. So by focusing your efforts on one or two or three layers, you don't have to concern yourself with the other layers. And in terms of communication, any one layer is in fact only aware of the layer above it. So the application layer, for example, has really only the user above it. And so it only needs to be aware of the presentation layer.

Similarly, the physical, there's nothing below it, so it only needs to be aware of the data link layer. 

But some of the ones in the middle, such as the transport layer, need to be able to inter-operate with the session layer and with the network layer. But still, nothing beyond that. So if I'm developing a protocol to transport layer, my world is the session layer, the transport layer, and the network layer, and that's it. 

Now, we'll talk about what each layer does in the upcoming presentations.But for the time being, suffice to say, that as you begin your communications, the packet begins at the top. The application layer on the sending side, and then travels down the stack to the bottom. And each layer does whatever it's supposed to do to that packet to get it into a format to be transferred across the physical wire. So once it's ready to go, it is then transmitted and accepted by the bottom physical layer on the receiving side. So it travels down the stack and across the wire and then the physical on the other side picks it up, if you will. And then the packet travels up the stack on the receiving side, and ultimately arriving at the top, the application layer of the recipient. So, this is what is looks like. We see the sender on the left. And again, imagine the sender is me, and I'm sending you an email. So I'm using some kind of application to type my message. And the application layer accepts that and sends it down the stack. Each layer does what it needs to do to the packet, it travels across the wire. You, as the recipient, will see that on the wire, and you will pick it up. Now it's at your physical layer, and it sends it up the stack. And whatever was done to it on the sending side is, for lack of a better word, undone on the recipient side.So that by the time your application receives that packet, it's the exact same data that I sent in the first place. 

Now to give you one very quick example of what that looks like the presentation layer is responsible for things like compression and encryption. So, and again, we'll talk about this later, but if I was sending you an email, and I wanted it to be safe, I might digitally sign that email and encrypt it. It's the presentation layer responsible for applying the encryption, sends it down the stack. You receive it, and it's the presentation layer on your side that's responsible for decrypting it and presenting you with the data that I sent originally. So that it's in the readable format, so you can understand it. If the presentation layer on your side did not do its job, you'd be presented with encrypted data that made no sense. So that's basically this virtual communication between layers. The presentation layer on my side inter-operates with the presentation layer on your side. And again, that happens at every layer. There's a communication between adjacent protocols. Each layer does something to the packet, which is then supposed to be handed by the corresponding layer on the recipient side. So again, that's what it means when it says virtual communication between layers. Any given layer is only concerned with what was done to it on the sending side of the same layer, not really anything else.

So again, as a developer, it's basically up to you to decide where you want to kind of work within that stack if you will. You generally don't find people that can develop for the entire stack. I'm not saying it's impossible, by any means, you may be a very skilled developer and you can work in all of them, but you typically tend to specialize. If I'm going to be an application developer, I'm just not going to concern myself with building new routers and switches and things along those lines that operate down at lower layers. And conversely, if I'm concerned with developing hardware, I'm not going to concern myself with developing applications. So again, this allows for that consistency. Everybody follows the same approach, the same plan. And we can rest assured that, if we do follow this approach, then everything should be ultimately compatible in terms of supporting network communications.

The term for the information being communicated at common levels of the OSI model is a PDU or protocol data unit. The applications information, presented at layers 5 through 7, is called data. The transport layer, headers and data that it contains, are called segments. And that's because they are used to divide up the original data into network byte-sized pieces. At the network layer, we refer to the header as a packet. And the protocol data unit for the data link layer is referred to as a frame. Because the data has a header on one end, and a trailer on the other, meaning the data has been framed!

[The Protocol Data Unit for Layer 1, Physical, is known as Bits.] To put it all together, HTTP data is placed within TCP segments, which are addressed with IP packets. Which are then put on the wire using ethernet frames. 

Now don't confuse PDU with MTU, or Maximum Transmission Unit. [The largest number of bits that will be interpreted as a frame on a particular link. Segmentation: Layer 4 division of data to meet MTU - guaranteed. Fragmentation: Layer 3 division of data to meet MTU - not guaranteed.] This is the byte-size piece that a layer 2 network can support. Ethernet, for example, commonly supports carrying 1,500 bytes of data in its frame. If a frame is received on one interface of a router, but must exit another interface, using a smaller MTU. The packet will be fragmented at layer 3 before re-transmission. In this example we can see that data moving from one system to another would not have any problems because this is all under the MTU. So that the original payload of 3.600 bytes will be divided into 1,500 byte chunks and sent on its way and get received on the other end. But if that same file was sent the other way, when it reaches a router that accepts 9,000 on one side but only 1,500 on the other. We will see fragmentation occur in order to make this particular block of data acceptable for the network devices that are in that network. The challenge with fragmentation is that there's no guarantee that the data will get there. So if you lose one fragment, you'll have to resend all the fragments that made up the original segment.

###### 2. Layer 1 - Physical
***

Now let's begin taking a closer look at the OSI model layer by layer. And, in this presentation, we'll start with, of course, Layer 1 or the physical layer. This is responsible for transmission of digital data bits from the source device to the destination device over the physical network communications media. So, as its name implies  at this layer  we find physical devices, hardware. So there are four main functions happening here. And it begins with the definition of hardware specifications, including the details for cables, connectors, network interface cards, repeaters, hubs. So what you might notice about the devices at this layer is that there is very little, if any, programming going on. You don't configure a cable. You don't go in and program a connector. You basically just physically set them up and they work. So really no configuration, nothing has to be accessed in terms of an interface to get to it, it's just there. It's ready to go. So those are the types of components that we find at the physical layer. Now the network interface card is a little bit of an exception. We know that there is some programming that's built into a network interface card and, of course, it has a driver for the operating system. But, from the perspective of configuring a network, you just insert the network interface card into the computer or of course it's already built in.

And again, it works, it's already set to do what it needs to do. And it is, after all, a hardware component so it does correctly reside at the physical layer. Then we see encoding and signaling. This is used to transform data from bits, 1s and 0s, into electrical signals. Those signals are then sent over the network. Then we see topology and physical network design. Then we see data transmission, which is actually sending and receiving the data. And, of course, for the particular type of network, for example, is it wired or wireless. Because, of course, there are different methods for each one. Then we see topology and physical network design. This considers things such as LAN  local area networks  versus WAN  wide area networks. Again, there are different considerations, different methods for sending the data when you have those different types of topologies. So another component that you'll also find at the physical layer is the bit rate control, which is the transmission rate or the number of bits per second.

So that you don't overwhelm the physical media by trying to send too much. So, again, we see the 1s and 0s that come down from the data link layer, which is Layer 2.

Then at the physical layer, they need to be converted into electrical signals that can travel the physical transmission medium. They are then received on the other side by the physical layer, they're converted from those electrical impulses back into 1s and 0s, and sent up to the data link layer on the receiving side. So again, each layer communicates only with its counterpart layer on the other side. But, ultimately, at the physical layer  again as mentioned  you don't see a whole lot of programming or configuration. Everything is very basic, very fundamental. But, of course, it is where, really, everything actually happens. The only way to get the data from one point to another is for it to cross that physical medium. And it is the responsibility of Layer 1  or the physical layer  to handle that portion of the communication. So it is of course the beginning of all network communication.

###### 3. Layer 2 - Data Link
***

Now, in this presentation, we'll take a look at Layer 2, or the data link layer. And this is responsible for node-to-node data delivery. Now, what that effectively means is that this is where we start to see the specific identification of a sender and a receiver. The physical layer really is not aware of that. It is just responsible for 1s and 0s and handling the transmission of data across the physical medium. So Layer 2 receives data coming down from the network layer and then passes it on to the physical layer. But during that process, it creates frames and adds physical address. Now very loosely, frames, in essence, identify the beginning and the end of any given transmission. Just like a frame around a picture outlines or borders the picture, if you will. But, really, it's the idea of the address that comes into play here. Now, it is known as a physical address and we'll also see in a moment a term that you might be familiar with, a MAC address, a Media Access Control address, but this is the first instance of any kind of address. There is no addressing at the physical layer.

So now that we do see an address appearing, this is how we can define who is the sender, who is the receiver. Now it also handles error-free transfers. So the other layers can assume error-free transmission over the medium. Now, we've talked already about different types of implementations where acknowledgements are used to say something was wrong with that, please resend it. But this, again, is at a much lower level. So we're talking about 1s and 0s, really, at this point, being missed. So the data link layer, in essence, is still responsible for saying something's not right here, this needs to be re-transmitted. So it's a very similar process, of course, but just at a more fundamental level. And it does this by implementing what's known as a checksum. So a checksum is a mathematical calculation that is applied to a chunk of data that says if you have this value and that value, then it should add up to this other value. So this is a very, very loose description. But let's just say you are sending the number 2, then you can say well, let's add to that number and you should get an answer of 4. That's a very simplistic example of a checksum. So if I receive this unit of information, and let's just say one of the values was lost so I only get 1 as a value and I'm told to add to that value. I will get an answer of 3, but I'm told the answer is supposed to be 4. Therefore, I can assume I missed a value. I did not get 2, I only got 1. Therefore, something was wrong. So please, send another value. Again, a very simple example of a checksum. But then we also see the source and the destination address come into play here and, of course, the data itself. So we're starting to see this definition of format characteristics of the transmission itself that just don't happen at the physical layer.

[Data Link Layer Functions]

So there are several other functions happening at this layer. Physical addressing, which we did talk about and, again, that is also known as the MAC address. But frame traffic control, frame sequencing, frame acknowledgement, again, detecting and recovering errors occurring at the physical layer. Frame delimiting, again, identifying the start and the end. And frame error checking, that checksum process. So we're starting to see much more control over the transmission itself, as opposed to just the physical layer, where it's just sending out a bunch of 1s and 0s.

We also see establishing and terminating the links so we know when to open the communication and close it. Media access management, meaning who is actually using the media at this point in time? Flow control, error control, and access control. These are all implemented at the data link layer to just implement some form of management to the communications. Because if there's no management, then basically everyone is going to try to access the media at the same time. And there's just no way to control that type of traffic, so the data link layer starts to implement these types of control mechanisms.

[Logical Link Control (LLC)]

Now, as the OSI Model developed, the data link layer was subsequently divided into two sublayers, or subcomponents, if you will. The LLC, which is the Logical Link Control sublayer, and the MAC sublayer, or Media Access Control. So looking first at the LLC, this provides an interface between the media access methods, which are down at the physical, and the network layer protocols, which are above it in Layer 3. So it's just a way to pass data from one to the other. But it also determines, at this point, if the communication is going to be connectionless or connection-oriented. Now that is essentially at the request of higher level protocols such as TCP or UDP up in the transport layer, but this is what you see the actual implementation. Some of those rules that we defined for connection-oriented communications are implemented by the LLC.

The Media Access Control, or the MAC, sublayer is responsible for the connection to the physical media. So actually getting it down to the 1s and 0s. But before it does, it applies the MAC address, a 12 digit hexadecimal number which is unique to every single computer, or more specifically stated, every single network interface in the world. There is never a duplicate of a MAC address. So the LLC, not concerned with the addressing, more so the format, the control mechanisms. The MAC, not concerned with those mechanisms, it's more concerned with the addressing. But both of them put together define the overall data link layer.

###### 4. Layer 3 - Network
***

Now, our next layer is Layer 3, or the network layer. And this is responsible for packet addressing, which I'll come back to in a moment, because we have touched on addressing already. But more to the point, we see a conversion of addresses happening here, logical to physical. Now, that actually depends on the direction. If you are the sending system, then a packet is travelling down the stack. And a logical network address does get passed to the data link layer, where we find the physical, or MAC, address. But if you are at the receiving side, the pack is travelling up the stack in the opposite direction. So a physical MAC address is then converted to a logical IP address. And therein lies the distinction. Your physical, or MAC, address is always the same, it never changes. The logical address, the IP address, can be changed at any time. And the IP address resides entirely at the network layer. And if you have ever configured any kind of network interface, you know that you can apply any valid address that you want. If you are in different locations, for example, even a small home network, you might be familiar with the fact that almost all of them use the same 192.168.something.something logical IP address.

In other words, many of them are reusable. You can find the exact same address in one network as you might in another. But it's because they are separated, they are different networks, that you can reuse those same IP addresses over and over again. So again, logical are a little more flexible. They're reusable and they can be changed. Physical cannot, they will always remain the same. So now we start to see a source to destination delivery. But using a particular method, and in this case, that method is routing. So getting packets from network to network, which is of course absolutely required to use the Internet. You would not have to use routing if all of the systems are in the same network. And in fact, you wouldn't even need a logical IP address, because the MAC address would be sufficient. Every system within that collection would be able to distinguish each other by their unique MAC addresses. But as soon as you have two separate networks, now you need a way to connect those networks, and that is routing. So routing happens entirely at the network layer, and as such, requires a logical IP address to be able to implement routing.

So like all layers, the network layer takes services from the data link layer, the layer below it. And then provides services to the transport layer, the layer above it. And again, this is dependent on the direction. So we're travelling up the stack, in this case. But that's fine, it just underscores the idea that any given layer is only aware of the layer above it and the layer below it. That's its whole world. So if we were on the sending side, it would be the exact opposite. It would receive from the transport and send to the data link. But again, it's just a fact that it only knows about those two layers.

So in terms of the functions, you see things such as subnet control within any given subnet that is bordered by a router. It can implement control such as throttling. It can say, hold on a second I'm too busy here, things are really filling up in my buffers, I'm accepting too much data. So the other sides need to slow down. Subnet usage accounting can keep track of frames, if you are in an environment where billing is an issue based on overall bandwidth usage. You can produce those types of accounting invoices to say, you used this much in terms of bandwidth. And again, that address mapping, the translation of logical addresses into physical addresses. Because they both still need to be known, but the network layer is responsible for applying the logical address.

Internetworking, of course, provides the capability for one network to communicate with another one. So it provides that logical connection between different networks. And again, the Internet is the best example of that. And we can also combine various networks to form a larger network. Within fairly large environments, they will typically subnet their networks into smaller units. But at the same time, those smaller units combine to form a larger network. So logical addressing, again, combines a large number of networks. It allows you to create a very large environment, up to the size of the Internet. But has to define an addressing scheme to still be able to uniquely identify each device on the internetwork. So again, if you think about the large organization, never mind the Internet, you can still define an internetwork. If you have, let's say, several physical or geographical locations, one in this city, one in that city, and one in another city. Well, you can connect all three of those and make your own internetwork. So again, you're taking network one and combining it with network two, and combining it with network three. That is an internetwork, and you need to be able to get between them. And yet, you still need to accommodate each individual device on each internetwork section. So you still have to accommodate everyone, and this is where the addressing scheme comes into play. Now, we will talk about that in much greater detail later on, as well.

We also see packetizing, which creates packets upon receiving data from upper layers. The packets are created by way of encapsulation, which is basically surrounding a packet with more information on either side. It's almost like a packet within a packet. And the IP protocol, which, again, resides at the network layer, defines the packet format, what is the structure. And of course, the IP address is part of that format. The sending system and the recipient, both need to be defined as part of the packet format.

Fragmentation is also implemented at this layer. And fragmentation involves dividing larger packets into smaller fragments. In some communications, the packet might simply be too large. So IP will divide it up into smaller units and then identify each one, so that it can be more easily sent out over the medium. And then, because each one is identified, it is able to be reassembled once it has been received, even though it has been fragmented. So ultimately, the network layer is primarily responsible for defining networks. It's not so much concerned with the individual systems within a network, but rather how do I get data from network A, to network B, to network C? And again, the other primary component is, of course, the IP address. That logical address that is changeable, is reusable. And those other functions will always be implemented at this layer. But again, as its name indicates, the primary job of the network layer is to define the boundaries of any given network and how to pass data between those networks

###### 5. Layer 4 - Transport
***

Continuing on, our next layer is layer 4 or the transport layer. And here we find the management of end-to-end message delivery, from the source, right through to the destination. And along the way, providing error checking to ensure that there are no duplications or errors during the data transfer. Ultimately, so that messages arrive intact and in the correct order. Now, as far as the layer itself, it resides at the core of the OSI model. And as packets are traveling up the stack, the transport layer takes its services from the network layer below it, and then provides services to the application layers above it. 

And on the way down, as the packet travels down the stack, one of the primary functions is the segmentation of any given message. It accepts the message from the session layer, and again, passes it to the network layer. But what we see here is the message being divided up into packets. That's one of the primary functions of the transport layer.

Now some other functions include the message acknowledgment and traffic control. Now that does depend on the protocol that's being used, but when it is being implemented this is where it happens. We also see what's known as session multiplexing. And this is due to the fact that any one system might be communicating with any number of other systems. So it multiplexes those connections together over one logical link. And it tracks which messages belong to which sessions. And it also implements what's known as service point addressing when you may still have just two systems. But there can be several applications in use at the same time that are all communicating with that other system. So service point addressing is to ensure that the correct packets arrive not only at the correct system, but are processed by the correct application.

Now, we also see flow control implemented here and we did see this down at the data link layer, but this is re-negotiated, if you will, using different protocols. And it prevents the source device from sending the data packets at a faster rate than the destination can handle. We also see error control to ensure that the entire message arrives without error.

Now again, this is where the two transport protocols of TCP and UDP reside. Again, recall that TCP is connection-oriented, UDP is not. So depending on the protocol you will see one of two things happen. So if you are using TCP then it is a connection-oriented transmission and this provides reliable transport of data. The receiving device sends an acknowledgement upon successfully receiving the packet. So we see the sender and the receiver implementing what's known as synchronization. And even that synchronization is acknowledged, meaning they are just setting things up. They are establishing the rules of communication and they are both agreeing on that. They basically negotiate how the transfer is going to take place. So that's what the synchronizations and the acknowledgements are all about. Once that is complete and both parties agree, then the connection is established. At that point we can start sending the data.

But in a connectionless transmission, something using the UDP protocol, the receiver does not acknowledge the receipt of a packet. The sending device simply assumes the packet arrived successfully. And again, this is not as reliable, but it enables faster communication between the devices. So ultimately, again, it really comes down to what you need. Do you need reliability or do you need speed? And you simply choose the appropriate protocol. But if it is UDP it will be connectionless, and we can get a little better performance, perhaps not as reliable but sometimes you just want the speed. So all of this is happening at the transport layer and again ultimately as its name indicates it is the actual movement of the data. This is really where you see it getting from point A to point B, and it's just a matter of how that movement takes place.

###### 6. Layer 5 - Session
***

Moving along, our next layer is Layer 5, or the session layer. And this is responsible for communication between two devices. And that statement in and of itself might seem somewhat generalized and possibly even redundant. But in this context, it refers to the beginning, the maintenance, and the ending of any given communication. So let's go back to Layer 4 for a moment, the transport layer. We talked about establishing a connection at this layer. The connection is simply saying, all right, we have connected to each other. But that's just the beginning of the process. Again, think about a phone call. If I pick up the phone and dial your number, and you answer and say hello, there we have a connection. The session more so refers to the entire conversation. So the beginning is certainly establishing the connection. But maybe at some point during our conversation, one of us will get cut off or will lose signal. Well, we know that somebody should try to call the other person back. That's the maintenance. And then, once we both had said goodbye and the conversation is over, we know that we can hang up. We should not just leave the phone sitting there still connected. So that's the ending. So it's the overall communication, all aspects from beginning to end, that represents the session. So this ultimately regulates the flow of data between the two devices much more effectively. Now, the session protocol in use will define various parameters for the connections. But it is charged with managing the transfer of data, in terms of who can transfer and for how long. So again, you want to take turns basically, and you generally do not want to leave the connection open once the communication has completed. It should be terminated.

So some other functions include the session establishment, the maintenance and the termination. Which, again, enables the two different machines to establish a session. And we see here session support, which implements security on the session, name recognition of the users on these machines, and things like logging in, so passing credentials back and forth. So those are just all examples of things that should be implemented with a connection, some sort of negotiation should occur. And once it has all completed, everything should be just shut down. So we see, more so just a lot of synchronizations happening back and forth here at the session layer. It's not so much concerned with passing the data, but agreeing on how we will control the passing of that data.

And one of those aspects is what's known as dialog control. This determines which device communicates first and how much data will be sent. Now, with respect to dialogue control, there are three types. Simplex, half duplex, and full duplex. Now, simplex is one-way communication, one at a time. You can equate that to two-way radios, walkie-talkies, for example. If I push my button and speak, I cannot listen during that time. And if you are receiving, you cannot speak during that time. When you push your button, it cuts off the receiver. So you can only send or only receive, and it has to be one person at a time. Half-duplex is both ways, so bi-directional but still one at a time. Full-duplex is bi-directional, two at a time. Now again, to imagine that, a phone call kind of borders both of those, if you will. In theory, a phone call is full duplex because we can both speak and listen at the same time, bi-directional. Both of us could speak if we wanted to, but it wouldn't be a very effective communication. I wouldn't understand you, you wouldn't understand me. So again, in theory, it is full duplex because it has that capability, but in practice, half duplex would be a normal phone conversation. I speak, you listen, you speak, I listen, but it is bi-directional. So again, it really comes down to the type of protocol, the type of application. But in terms of network communications, full duplex is certainly the fastest. Both of us can send and receive at the same time.

So some other functions include dialog separation. This adds checkpoints or markers within a message, so that we know where the beginning and where the end is. And also where it might need to be resumed from if it was interrupted. Again, that's the maintenance. So some examples of the protocols that operate at the session layer, NetBIOS, named pipes, and RPC. But again, the overall idea of the session layer is just that maintenance that set of guidelines over the entire communication. So that for the most part, we just know when it starts, how to maintain it, and when it ends.

###### 7. Layer 6 - Presentation

***

Now, our next layer is Layer 6, or the presentation layer. Which is often refered to as the translation layer. And I'll come back to that point in a moment. But its primary job is to present data in a uniform format to the application layer. Which essentially means it masks any difference in data between two different systems.

So coming back to the translation, it formats the data that will be presented to the application layer. By translating that data into a common format, known to the application layer at the receiving station. So to visualize this, imagine two email applications. I'm just sending you an email, but I'm using email application X and you are using email application Y. And we have them set to use different data formats. Maybe I'm sending out in plain text and you're receiving in HTML. Well, in this case, it is the job of the presentation layer to ensure that the application does not need to concern itself with converting those types of data. The application simply wants to present the data it receives. But it's the presentation layer that is responsible for ensuring the application understands that data. And in essence, just does not have to worry about it. So this is where you see this translation happening. And that's the fairly simple example. We'll see some additional ones in a moment.

But we also see functions of character code translation happening here. ASCII, or American Standard Code for Information Interchange. And what's commonly referred to as EBCDIC which is Extended Binary Coded Decimal Interchange Code. These essentially are translation methods from binary to character information. They're just different approaches, but of course binary is at the lowest levels and character is what we expect to see. It also handles data conversion, with respect to the bit order. The CR-CR/LF stands for carriage return, and carriage return line feed. Carriage return is, off course, the old typewriter, when you hit Enter, and move the carriage back to a starting position. Its understanding and/or recognizing where those have happened, and line feed is more or less the same thing for Linux and UNIX systems. And integer floating point is recognizing the differences between whole integers, those without decimal places, versus floating points, those that do. So again, we see conversions happening here, from binary types of data and/or various formats of data into what the application can understand as character based data. That, of course, we ultimately need to understand.

So here are two of the most common functions that also occur at the presentation layer. And again, this comes back to the translation. Data compression is heavily used in just about every computing environment. Which, of course, reduces the number of bits that need to be transmitted over the network and/or stored on any kind of storage media. But compression is, in essence, a translation. It uses, in some cases, various sets of characters that represent other sets of characters. So for example, just to use a very simple one, you might use the symbol of the asterisk to represent the word hello. Well, a single character of the asterisk is, of course, much less data than the five characters in the word hello. So the data is compressed. But we need to know how to un-compress that data. We need to know the conversion.

So again, we translate one character to another. So again, it's the job of the presentation layer to do that. And data encryption is perhaps the best example of all, because encryption, typically the opposite in terms of size when it comes to compression. When you encrypt data, it typically is much larger,. But it's much safer because it scrambles characters and uses way more than what are actually required. So the word hello might be represented by a string of characters that makes no sense to us and could be possibly hundreds of characters in length. But that's the idea, we want to favor the security of that.

[By using passwords.]

But we still need a way to understand the data. We need to decrypt it.

So again, it is the presentation layer where this happens. If the presentation layer didn't exist, we would see a bunch of encrypted data in the application that makes no sense to us. If it was compressed, we'd only see a bunch of meaningless characters. So again, the presentation layer undoes the compression, undoes the encryption, and presents it to the application so that we see normal data. So again, it's something that really is responsible for translation, hence the term. But presentation layer is where this occurs so that the application itself just does not need to concern itself with handling all those different types of possibilities in terms of how the data is represented.

###### 8. Layer 7 - Application
***

All right, our final layer of the OSI model is layer seven, or the application layer, and this serves as the window for users and application processes that need to access other network services. It is the interface between the program and the rest of the protocol stack. So it's appropriately named, because as users, we directly interact with an application, but the program itself, the application, doesn't really reside within layer seven. If you will, it would be just above layer seven. That's what we use, as users, to interact with the system, but it is the protocols within layer seven that directly support that application and allow it to access the appropriate network services.

So to give you a better idea of what that means, these are some of the network services, and these are just a few of many. But FTP, File Transfer Protocol, DHCP, Dynamic Host Configuration Protocol, DNS, Domain Name System, SMTP, Simple Mail Transfer Protocol, HTTP, Hyper Text Transfer Protocol, POP3, Post Office Protocol version 3, and NFS, the Network File System, are all different types of network services. If I want an IP address, I use the DHCP network service to request an IP address, and I can use an interface, within my operating system to do so. So there is an application in front of me, or I can say renew this IP address, that is the application, but it is the DHCP protocol that handles the request to get that IP address.

A much more familiar example, open up a web browser. The browser itself is the program, and again, resides just above layer seven, but if you look at the address, you see, more often than not, HTTP://, then the name of the website. So that application, the browser, is directly dependent on the HTTP protocol to request the website from whichever server hosts that site. So again, that's the network service, getting the site. The browser is the program. So they entirely depend on each other, but it is the function of these application layer protocols to simply provide the services to the various types of applications that we use. 

So in terms of the functionality, it allows us, users, to interact with the applications. The application accepts user input and passes data down to the lower layers. Another example of that, using email. I just type in a message into the email interface. The application accepts that and sends it on its way down the stack. So this allows for easier application compatibility and implementation, and I'll clarify that in greater detail in a moment. But essentially, applications do not have to be re-written for different types of network environments.

So if we look at some examples, mail services, something we use every day. Email applications do not need to be consistent to send or receive messages across different environments. I don't need to concern myself with what type of email application you are using if I'm sending you an email. I don't need to concern myself with what type of network environment you are using. It's just a matter of me knowing your address, so I take advantage of the SMTP network service to send that email across the Internet, and in essence, you receive it. It doesn't matter which program you are using, nor does it matter what type of network environment you are in. File transfer, access, and management, or FTAM, multiple types of FTAM applications can be used to access and modify files remotely. FTP is a good example of that. If I'm a web developer, there's a very good chance that I develop my websites locally, and I transfer the files up to a server. So again, there are many different types of applications, and I don't need to concern myself with which one should I use. As long as it supports the FTP protocol, then I can use that application.

And of course, browsers, this provide Internet access for all types of networking environments. And as a developer, I don't need to create a browser for network environment a or network environment b. As long as the HTTP protocol can be accessed, that is really all I need to concern myself with. So that's where that compatibility and easier implementation comes into effect. I just don't need to concern myself with the lower layer protocols, as an application layer developer. As long as I use those standard protocols, then my application will work. So that's where you see that easier implementation. The compatibility just really is not a concern. As long as it's compatible with the application layer protocols, yes, that's a concern, but that is really as far as I need to go. I don't need to make different applications for different types of network interface adapters way down at the physical layer, or a different browser, one for the TCP protocol and one for the UDP protocol. As long as it can just access its immediate layer below the application, that's it. Everything else is up to the lower layers, and compatibility and integration and implementation becomes much easier to facilitate.
